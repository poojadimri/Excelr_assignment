```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Pooja/Desktop/EastWestAirlines.csv")
df
```

```python
df.info()
```

```python
# Normalization function
def standardize(x):
    x = (x-x.mean())/((x.std()))
    return x
df_scaled = standardize(df.iloc[:,1:])
df_scaled
```

```python
def minmax(x):
    x = (x-x.min())/(x.max()-x.min())
    return x
df_norm = minmax(df.iloc[:,1:])
df_norm
```

```python
# Hierarchical clustering
```

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
```

```python
dendrogram = sch.dendrogram(sch.linkage(df_norm,method = 'complete'))
```

```python
hc = AgglomerativeClustering(n_clusters = 5,affinity = 'euclidean',linkage = 'complete')
y_hc = hc.fit_predict(df_norm)
```

```python
clusters = pd.DataFrame(y_hc, columns = ['clusters'])
```

```python
clusters
```

```python
metrics.silhouette_score(df_scaled,y_hc)
```

```python
data_HC=df.copy()
```

```python
data_HC['clusters']=y_hc
data_HC
```

```python
data_HC.groupby("clusters").agg(["mean"]).reset_index()
```

```python
for i in range(0,11):
    sns.barplot(y=data_HC.iloc[:,i],x=data_HC["clusters"])
    plt.xlabel("Clusters")
    plt.ylabel(data_HC.columns[i])
    plt.title(f"hierarchical Clustering")
    plt.show()
```

```python
# KMeans Clustering
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_df=scaler.fit_transform(df.iloc[:,1:])
scaled_df
```

```python
wcss=[]
for i in range (1,11):
    kmeans=KMeans(n_clusters=i,random_state=0)
    kmeans.fit(scaled_df)
    wcss.append(kmeans.inertia_)
    
plt.plot(range(1,11),wcss)
plt.ylabel('wcss')
plt.title('elbow method')
plt.xlabel('number of clusters')
plt.show()
```

```python
clusters_new=KMeans(5,random_state=42)
clusters_new.fit(scaled_df)
```

```python
data_KM=df.copy()
data_KM["KMeans_n5"]=clusters_new.labels_
data_KM
```

```python
clusters_new.cluster_centers_
```

```python
data_KM.groupby('KMeans_n5').agg(["mean"]).reset_index()
```

```python
data_KM['KMeans_n5'].value_counts()
```

```python
for i in range(0,11):
    plt.figure(figsize=(12, 7))  
    sns.barplot(y=data_KM.iloc[:,i],x=data_KM["KMeans_n5"])
    plt.title(f"kmeans clustering")
    plt.ylabel(data_KM.columns[i])
    plt.xlabel('kmeans')
    plt.show()
```

```python
# DBSCAN
```

```python
from sklearn import metrics
from sklearn.cluster import DBSCAN
```

```python
dbscan = DBSCAN(eps=5,min_samples=13)
dbscan.fit(scaled_df)
```

```python
data_DB=df.copy()
data_DB['clusters']=dbscan.labels_
data_DB
```

```python
data_DB.groupby('clusters').agg(['mean']).reset_index()
```

```python
metrics.silhouette_score(scaled_df,dbscan.labels_)
```

```python
data_DB['clusters'].value_counts()
```

```python
for i in range(0,12):
    sns.barplot(y=data_DB.iloc[:,i],x=data_DB["clusters"])
    plt.xlabel("DBSCAN Clusters")
    plt.ylabel(data_DB.columns[i])
    plt.title(f"DBSCAN Clustering")
    plt.show()
```
