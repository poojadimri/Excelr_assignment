```python
data_y=wine.iloc[:,:1]
data=wine.iloc[:,1:]
data
```

```python
# normalizing the numerical data
std_df=scale(data)
std_df
```

```python
pca=PCA(n_components=13)
pca_std_df=pca.fit_transform(std_df)
pca_std_df
```

```python
pca.components_
```

```python
# the amount of variance that each PCA explains is
var=pca.explained_variance_ratio_
var
```

```python
# cumulative varience
var1 = np.cumsum(np.round(var,decimals = 4)*100)
var1
```

```python
pc=range(1,pca.n_components_+1)
plt.bar(pc,var,color='blue')
```

```python
# variance plot for PCA components obtained
plt.plot(var1,color='blue')
```

```python
pca_df=pd.DataFrame(pca_std_df[:,0:3],columns=['PCA 1','PCA 2','PCA 3'])
pca_df
```

```python
pca_df = pd.DataFrame(pca_std_df)
```

```python
plt.scatter(pca_df[0], pca_df[1], alpha=.3, color='blue')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.zlabel('PCA 3')
plt.show()
```

```python
#optimal number of clusters
```

```python
wcss = []
for i in range(1, 15):
    kmeans = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans.fit( pca_df.iloc[:,:3])
    wcss.append(kmeans.inertia_)
```

```python
plt.plot(range(1, 15), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
# from the graph k = 3 
```

```python
# K clusters
```

```python
model = KMeans(n_clusters=3)
model.fit(pca_df.iloc[:,:2])
```

```python
KMeans(algorithm='auto', copy_x=True, init='k-means++', max_iter=300,
       n_clusters=3, n_init=10, n_jobs=None, precompute_distances='auto',
       random_state=None, tol=0.0001, verbose=0)
labels = model.predict(pca_df.iloc[:,:2])
plt.scatter(pca_df[0], pca_df[1], c=labels)
plt.show()
```

```python
kmeans_cluster=pd.DataFrame(model.labels_,columns=['kmeans'])
```

```python
kmeans_cluster
```

```python
kmeans_df=pd.concat([data,kmeans_cluster],axis=1)
kmeans_df
```

```python
kmeans_df.groupby("kmeans").agg(['mean']).reset_index()
```

```python
# hierachy
```

```python
dendrogram=sch.dendrogram(sch.linkage(pca_df,method='ward'))
```

```python
hc=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage="ward")
```

```python
y_hc=hc.fit_predict(pca_df)
```

```python
hierarchy=pd.DataFrame(y_hc,columns=['h_clusters'])
hierarchy
```

```python
hierachy_clusters=pd.concat([data,hierarchy],axis=1)
hierachy_clusters
```

```python
hierachy_clusters.groupby('h_clusters').agg(['mean']).reset_index()
```
